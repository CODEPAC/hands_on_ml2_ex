```python
import os
from urllib import request
```

### fecthing Titanic data from url

```python
TITANIC_DATASET_PATH = os.path.join("datasets", "titanic")
DOWNLOAD_URL = "https://raw.githubusercontent.com/ageron/handson-ml2/master/datasets/titanic/"
```

```python
def fetch_titanic_dataset(download_url:str, dataset_path:str)-> None:
    if not os.path.isdir(dataset_path):
        os.makedirs(dataset_path)
    for filename in ["train.csv", "test.csv"]:# "gender_submission.csv"]:
        file_path = os.path.join(dataset_path, filename)
        if not os.path.isfile(file_path):
            print(f"downloading {filename}")
            request.urlretrieve(download_url + filename, file_path)
    
```

```python
fetch_titanic_dataset(DOWNLOAD_URL, TITANIC_DATASET_PATH)
```

```python
import pandas as pd
def load_csv_data(folder_path:str, filename:str)-> pd.DataFrame:
    file_path = os.path.join(folder_path, filename)
    data = pd.read_csv(file_path)
    return data
```

```python
train_data = load_csv_data(TITANIC_DATASET_PATH, filename="train.csv")
test_data  = load_csv_data(TITANIC_DATASET_PATH, filename="test.csv")
```
